Here are the steps for implementing the simple calculator program:

1. **Accept User Input:**
   - Prompt the user to enter two numbers.
   - Ensure the input is formatted to remove extra spaces and handle cases (upper/lower).

2. **Display Menu of Options:**
   - Present a menu of operations: addition, subtraction, multiplication, division, integer division, and exponentiation.
   - Ask the user to input their desired operation.

3. **Perform Valid Operation:**
   - Check if the entered operation is valid (add, subtract, multiply, divide, integer division, exponent).
   - If the operation is valid, proceed; otherwise, inform the user that the operation entered is not valid.

4. **Perform Calculation:**
   - Based on the chosen operation:
     - Perform addition if the user chose 'add'.
     - Perform subtraction if the user chose 'subtract'.
     - Perform multiplication if the user chose 'multiply'.
     - Perform division if the user chose 'divide'.
     - Perform integer division if the user chose 'integer division'.
     - Perform exponentiation if the user chose 'exponent'.

5. **Display Result:**
   - Show the calculated result of the chosen operation on the entered numbers.
   - Output the result to the user.

6. **Handle Errors:**
   - Ensure the program handles input errors gracefully, providing appropriate messages if the input format is incorrect or if an invalid operation is selected.

These steps outline the process of building the calculator program, ensuring user input is validated, operations are performed based on the chosen operation, and the calculated result is displayed. Implementing these steps in Python will result in a functional simple calculator program.

```python
# write your code here ...
while True:
    ######   app inputs and input validation:
    while True:
        num1 = input('Enter your 1st number\n')
        if not num1.isdigit():
            print('ERROR!, Enter a valid number')
            continue
        num2 = input('Enter your 2nd number\n')
        if not num2.isdigit():
            print('ERROR!, Enter a valid number')
            continue
        while True:
            opr = input('Select your desired operator from\n* add\n* subtract\n* multiply\n* divide\n* integer division\n* exponent\n')
            opr = opr.strip().lower()
            opr_list = ['add','subtract','multiply','divide','integer division','exponent']
            if not opr in opr_list:
                print('ERROR!, Enter a valid operator')
                continue
            elif opr in opr_list:
                break
        break
    ######   app logic:
    result = 0 
    num1 = float(num1)
    num2 = float(num2)
    if opr == 'add':
        result = num1 + num2
        print('Your result is \n', result)
    elif opr == 'subtract':
        result = num1 - num2
        print('Your result is \n', result)
    elif opr == 'multiply':
        result = num1 * num2
        print('Your result is \n', result)
    elif opr == 'divide':
        result = num1 / num2
        print('Your result is \n', result)
    elif opr == 'integer division':
        result = int(num1 / num2)
        print('Your result is \n', result)
    elif opr == 'exponent':
        result = num1 ** num2
        print('Your result is \n', result)
    print('Thnaks for using our calculator app')
    ######   app loop logic:
    l2 = ['new','exit']
    loop = input('Enter "new" to make a new calculation , or "exit" to close the program \n')
    while True:
        if loop not in l2:
            print('ERROR!, kindly type "new" or "exit"')
            loop = input('Enter "new" to make a new calculation , or "exit" to close the program \n')
            continue
        else:
            break
    if loop == 'new':
        continue
    elif loop == 'exit':
        break
        
    with open('result.txt', 'a') as file:
        file.write(f'{num1}{opr}{num2} = {result}\n')
```
